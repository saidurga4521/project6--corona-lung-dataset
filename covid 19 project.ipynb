```python
#import all the librarys required
from tensorflow.keras.layers import Input,Lambda,Dense,Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img #it is used for data augmentation
import numpy as np
from glob import glob #to return all file paths that match a specific pattern
```

```python
#resize all the images to this
IMAGE_SIZE=[224,224]
```

```python
train_path='/content/drive/MyDrive/programs/datasets/covid 19 train'
test_path='/content/drive/MyDrive/programs/datasets/covid 19 test'
```

```python
# import resnet 50 library as shown below and add preprocessing layer to the format of resnet 
#here we will be using image nets
resnet=ResNet50(input_shape=IMAGE_SIZE+[3],weights='imagenet',include_top=False)


```

```python
#Don't train existing weights or dont train the base model
for layer in resnet.layers:
  layer.trainable=False


```

```python
#using for getting number of out put classes
folders=glob('/content/drive/MyDrive/programs/datasets/covid 19 train')
```

```python
print(folders)
```

```python
resnet.output
```

```python
#you can add the our required output layers
x=Flatten()(resnet.output)
```

```python
len(folders)
```

```python
#we have to add the dense layer
prediction=Dense(len(folders),activation='softmax')(x)
```

```python
#now create the model
model=Model(inputs=resnet.input,outputs=prediction)
```

```python
#now summary of the model
model.summary()
```

```python
# to compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#use the image data generator to import the images from image dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_data=ImageDataGenerator(rescale=1./225,
                              shear_range=0.2,
                              zoom_range=0.2,
                              horizontal_flip=True)
test_data=ImageDataGenerator(rescale=1./225)
```

```python
#now read the training images
training_set=train_data.flow_from_directory('/content/drive/MyDrive/programs/datasets/covid 19 train',
                                            target_size=(224,224),
                                            batch_size=32,
                                            class_mode='categorical')
 
```

```python
#now read the testing images
testing_set=test_data.flow_from_directory('/content/drive/MyDrive/programs/datasets/covid 19 test',
                                          batch_size=32,
                                          class_mode='categorical')
```

```python
#finally we get the test and train data
#now fit the data
model.fit(training_set,epochs=10)
```
